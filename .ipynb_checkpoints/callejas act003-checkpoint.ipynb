## Desafío 1: Preparar el ambiente de trabajo
- Importe los módulos numpy , pandas , matplotlib , seaborn , glob y os siguiendo las buenas prácticas. Los últimos dos módulos permitirán realizar la importación de múltiples archivos dentro de la carpeta dump .
- Para ello genere un objeto que guarde en una lista todos los archivos alojados en dump utilizando y para extraer las rutas absolutas. Posteriormente genere un objeto que contenga todos los csv.
- Asegúrese de eliminar la columna Unnamed: 0 que se genera por defecto.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob, os
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
```

```python
file_list = glob.glob(os.getcwd()+'/dump/*.csv')

append_csv = []
# para cada csv
for f in file_list:
    # abrir en pandas y concatenar
    append_csv.append(pd.read_csv(f, index_col=None, header=0).drop(columns='Unnamed: 0'))
# concatenar la lista de dataframes


df_lyrics = pd.concat(append_csv)
df_lyrics.columns = ['Artist', 'Genre', 'Song', 'Lyrics']    
```

## Desafío 3: Matriz de ocurrencias
- Importe la clase dentro de los módulos feature_extraction.text de la librería . Lea la documentación asociada a ésta. ¿Cuál es el objetivo de esta clase?
- Aplique la clase para extraer las 100 palabras más repetidas en toda la base de datos. Genere una función que replique el procedimiento para cada uno de los géneros.
- Comente sobre las principales características de cada género en cuanto a sus palabras.

```python
cv = CountVectorizer(stop_words='english')
```

```python
cv_fit = cv.fit_transform(df_lyrics['Lyrics'])
```

Extraemos tokens

```python


cv_fit = cv.fit_transform(df_lyrics['Lyrics'])
words = cv.get_feature_names()

word_freq = cv_fit.toarray().sum(axis=0)

words_df = pd.DataFrame({'words': words, 'freq': word_freq})

plt.figure(figsize=(25,10))
plt.xticks(rotation=60)
words_df.sort_values(by='freq', ascending=False, inplace=True)
sns.barplot(x='words', y='freq', data=words_df.head(100))
```

```python
for i in df_lyrics['Genre'].unique():
    
    df = df_lyrics[df_lyrics['Genre']==i]['Lyrics']
    cv_fit = cv.fit_transform(df)
    words = cv.get_feature_names()

    word_freq = cv_fit.toarray().sum(axis=0)

    words_df = pd.DataFrame({'words': words, 'freq': word_freq})

    plt.figure(figsize=(20,10))
    plt.title(i)
    plt.xticks(rotation=60)
    words_df.sort_values(by='freq', ascending=False, inplace=True)
    sns.barplot(x='words', y='freq', data=words_df.head(100))
```

- Las distribuciones de pop y rock son muy similares, ambas contienen love, like, yeah, time, come, want entre otras.

# Desafío 4: Entrenamiento del Modelo
   
- Importe  Pipeline , confusion_matrix y MultinomialNB train_test_split
- Genere las muestras de entrenamiento y validación reservando un 40% para validación y
declarando una semilla pseudoaleatoria.
- Monte el modelo dentro de un Pipeline , donde el primer paso es implementar
CountVectorizer y el segundo sea ejecutar el clasificador MultinomialNB .
- A continuación se les presenta una lista de letras, ¿cuáles serían las predicciones correspondientes?
- Genere una predicción implementando la muestra de validación y contraste las predicciones del modelo con las etiquetas verdaderas. 
- Reporte las principales métricas.

```python
X_train, X_test, y_train, y_test = train_test_split(df_lyrics['Lyrics'],
                                                  df_lyrics['Genre'],
                                                  test_size=.40,
                                                  random_state=3504)

text_classifier = Pipeline(
    [
        ('vectorizer', CountVectorizer(stop_words='english')),
        ('classifier', MultinomialNB())
    ]
)

text_classifier.fit(X_train, y_train)
```

```python
text_to_predict = ['I got a place in the underworld', # Brody Dalle -Underworld
                  'As veils of ignorance, hatred retains Storm of arrows through karma Seeking light through samsara', # Gorguts - Forgotten Arrows
                  "Bye bye Don't want to be a fool for you Just another player in your game for two You may hate me but it ain't no lie",
                   # N'SYNC - Bye Bye Bye
                  'Move bitch, get out the way Get out the way bitch, get out the way Move bitch, get out the way Get out the way bitch, get out the way', # Ludacris - Move B*tch
                  'Sexual violence doesn’t start and end with rape It starts in our books and behind our school gates' # IDLES - Mother,
                  "Take it from the girl you claimed to love You gonna get some bad karma I'm the one who had to learn to  build a heart made of armor From the girl who made you soup and tied your shoes when you were hurting You are not deserving, you are not deserving" #Banks - Drowning
]
```

```python
text_classifier.predict(text_to_predict)
```

*N'Sync es rock? WTF

```python
y_hat = text_classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_hat))
```

```python


cnf = confusion_matrix(y_test, y_hat)/len(y_test)

# guardamos las etiquetas de las clases
target_label = df_lyrics['Genre'].unique()
# Implementamos un mapa de calor definiendo las clases
sns.heatmap(cnf, xticklabels=target_label,
            yticklabels=target_label,
            # generamos las anotaciones en términos porcentuales
            annot=True, fmt=".1%", 
            # evitamos la barra y cambiamos el colormap
            cbar=False, cmap='Blues')
```

Revisando tanto el clasification report, como la matriz de confusión, notamos que *pop* es el género para cual el modelo funciona peor (f1-score de 0.47), de hecho, hay más canciones pop que son calificadas como rock, que bien calificadas como pop. Esto se puede deber que las clases están desbalanceadas

# Desafío 5: Mejora del Modelo
- Proponga una estrategia para mejorar el desempeño del modelo en la categoría con peores métricas.
- Repita los pasos de entrenamiento y reporte de métricas, esta vez incluyendo los nuevos datos suministrados.
- Comente sobre el desempeño general de éste.

Debido a que las clases están desbalanceadas, agregaremos más datos de pop, con la esperanza de que este nuevo modelo pueda mejorar las predicciones

```python


append_csv_pop = []
file_list = glob.glob(os.getcwd()+'/dump_plus_pop/*.csv')
for f in file_list:
    # abrir en pandas y concatenar
    append_csv_pop.append(pd.read_csv(f, index_col=None, header=0).drop(columns='Unnamed: 0'))
# concatenar la lista de dataframes

df_lyrics_pop = pd.concat(append_csv_pop)
df_lyrics_pop.columns = ['Artist', 'Genre', 'Song', 'Lyrics']  
```

```python
df_lyrics_pop.head()
```

```python
df_group_genre = df_lyrics_pop.groupby('Genre').count().reset_index()
df_group_genre.sort_values(by='Lyrics', ascending=False, inplace=True)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=45)
sns.barplot(x='Genre',y='Lyrics', data=df_group_genre)
```

De esta forma, pop tiene más datos y metal ahora es la clase con menor cantidad de datos

```python
X_train, X_test, y_train, y_test = train_test_split(df_lyrics_pop['Lyrics'],
                                                  df_lyrics_pop['Genre'],
                                                  test_size=.40,
                                                  random_state=3504)

text_classifier = Pipeline(
    [
        ('vectorizer', CountVectorizer(stop_words='english')),
        ('classifier', MultinomialNB())
    ]
)

text_classifier.fit(X_train, y_train)
```

```python
y_hat = text_classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_hat))
```

Mejora el F1 de pop y metal, al costo de disminuir el de los demás generos. De todos modos, el aumento del F1 score de pop es considerablemente mayor que la disminución del F1 de los demás

```python

cnf = confusion_matrix(y_test, y_hat)/len(y_test)

# guardamos las etiquetas de las clases
target_label = df_lyrics_pop['Genre'].unique()
sns.heatmap(cnf, xticklabels=target_label,
            yticklabels=target_label,
            # generamos las anotaciones en términos porcentuales
            annot=True, fmt=".1%", 
            # evitamos la barra y cambiamos el colormap
            cbar=False, cmap='Blues')
```

A diferencia del modelo anterior, ahora los casos correctamente predichos de pop superan a aquellos que son predichos erroneamente
